```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("istherecorrelation.csv", sep=";", decimal=",")
data.columns = ["year", "WOstu", "BeerConsump"]
```

```python
data
```

```python
# this fig is not good enough to show the correlation between WO students and the beer consumption
fig, ax = plt.subplots()
ax.plot(data.year,data.WOstu,label="WOstudents(x1000)")
ax.plot(data.year,data.BeerConsump/50, label="BeerConsumption(x50000)")
ax.legend()
plt.xlabel("Year",fontsize=11)
plt.ylabel("Number of People",fontsize=11)
plt.title("WO Student Numbers and Beer Consumption over the Years")
# plt.savefig("WOstu_year.jpg",dpi=300)
```
